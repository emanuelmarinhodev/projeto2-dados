# 4. Feature Engineering

```python
df_pedidos['Receita_Item'] = df_pedidos['Quantidade'] * df_pedidos['Preco_Unitario']
```

```python
df_pedidos['Receita_Item'].isnull().sum()
```

# 5. Tratamento de Dados

```python
df_pedidos.isnull().sum()
```

```python
df_pedidos_tratado = df_pedidos.copy()

df_pedidos_tratado['Quantidade'] = df_pedidos_tratado['Quantidade'].fillna(df_pedidos_tratado['Quantidade'].mean())
```

```python
df_pedidos_tratado = df_pedidos_tratado.dropna(subset=['Preco_Unitario'])
```

```python
df_pedidos_tratado.isnull().sum()
```

```python
df_pedidos_tratado['Receita_Item'] = df_pedidos_tratado['Quantidade'] * df_pedidos_tratado['Preco_Unitario']
```

# 6. Agregações

```python
df_pedidos_tratado.columns
```

```python
df_pedidos_tratado['Quantidade'] = df_pedidos_tratado['Quantidade'].astype(int)
```

```python
df_pedidos_tratado.info()
```

```python
quantidade_por_item = df_pedidos_tratado.groupby('Item')['Quantidade'].sum().reset_index()

quantidade_por_item
```

```python
receita_total_item = df_pedidos_tratado.groupby('Item')['Receita_Item'].sum().reset_index().rename(columns={'Receita_Item': 'Faturamento_Total'})
receita_total_item['Faturamento_Total'] = receita_total_item['Faturamento_Total'].round(2)
receita_total_item
```

```python
quantidade_por_item.sort_values(by='Quantidade', ascending=False).head(5)
```

```python
receita_total_item.sort_values(by='Faturamento_Total', ascending=False).head()
```

# 7. Análise Temporal

```python
df_pedidos_tratado.columns
```

```python
df_pedidos_tratado['Data'] = pd.to_datetime(df_pedidos_tratado['Data'], errors='coerce')

```

```python
df_pedidos_tratado.info()
```

```python
df_pedidos_tratado['Mes'] = df_pedidos_tratado['Data'].dt.month
```

```python
df_pedidos_tratado.head()
```

```python
receita_por_mes = df_pedidos_tratado.groupby('Mes')['Receita_Item'].sum().reset_index()

receita_por_mes
```

```python
evolucao_vendas_tempo = df_pedidos_tratado.sort_values('Data').groupby('Data')['Receita_Item'].sum().reset_index()
```

# 8. Merge


```python
df_cardapio = pd.read_csv('./dataset/cardapio.csv')

df_cardapio.head()
```

```python
df_cardapio.tail()
```

```python
df_cardapio.info()
```

```python
df_cardapio.describe()
```

```python
df_pedidos_tratado['Item'].nunique()
```

```python
df_merge = pd.merge(df_pedidos_tratado, df_cardapio, how='left')
```

```python
df_merge.columns
```

```python
receita_total_categoria = df_merge.groupby('Categoria')['Receita_Item'].sum().reset_index()

receita_total_categoria
```

```python
index_categoria_maior_receita = receita_total_categoria['Receita_Item'].idxmax()

print(f'A categoria com maior receita é: {receita_total_categoria['Categoria'][index_categoria_maior_receita]}')
```

# 10. KPIs

```python
df_merge.columns
```

```python
receita_total = df_merge['Receita_Item'].sum()
print(f'A receita total gerada foi: R${receita_total:.2f}')
```

```python
total_itens_vendidos = df_merge['Quantidade'].sum()
print(f'A quantidade total de itens vendidos foi: {total_itens_vendidos}')
```

```python
ticket_medio = receita_total / df_merge['ID_Pedido'].nunique()
print(f'O ticket médio foi: R${ticket_medio:.2f}')
```

```python
q1_preco, q2_preco, q3_preco = np.percentile(df_merge['Preco_Unitario'], [25, 50, 75])

print(f'25% preço: {q1_preco}')
print(f'50% preço: {q2_preco}')
print(f'75% preço: {q3_preco}')
```

```python
q1_qtd, q2_qtd, q3_qtd = np.percentile(df_merge['Quantidade'], [25,50,75])

print(f'25% quantidade: {q1_qtd}')
print(f'50% quantidade: {q2_qtd}')
print(f'75% quantidade: {q3_qtd}')
```
